```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = (8,8)
```

```python
from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    RobustScaler
)
```

```python
X = np.random.randint(0, 10, 100).reshape(-1, 1)
```

```python
X[:4]
```

```python
some = {
    'original': X,
    'minmax': MinMaxScaler().fit_transform(X),
    'standard': StandardScaler().fit_transform(X),
    'robust': RobustScaler().fit_transform(X)
}
```

```python
x = range(len(X))
colors = iter(['black', 'red', 'cyan', 'green'])
for name, scaled in some.items():
    plt.scatter(x, scaled, color=next(colors), label=name)
plt.legend()
```

```python
X = np.random.randint(0, 10, 100)
X1 = np.array([100, 130, 125, 140])
```

```python
X = np.concatenate((X, X1)).reshape(-1, 1)
plt.boxplot(X)
```

```python
some = {
    'minmax': MinMaxScaler().fit_transform(X),
    'standard': StandardScaler().fit_transform(X),
    'robust': RobustScaler().fit_transform(X)
}
```

```python
x = range(len(X))
colors = iter(['red', 'cyan', 'green'])
for name, scaled in some.items():
    plt.scatter(x, scaled, color=next(colors), label=name)
plt.legend()
```
